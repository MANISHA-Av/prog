```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
## Functional
plt.plot(x,y,'r')

plt.xlabel('x label')
plt.ylabel('y label')
plt.title('title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#object oriented
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('x label')
axes.set_ylabel('y label')
axes.set_title('set title')
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('larger plot')
axes2.plot(y,x)
axes2.set_title('smaller plot')
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.15,0.4,0.3])
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[0].set_title('first plot')


axes[1].plot(y,x)
axes[1].set_title('second plot')
```

## fig size and DPI

```python
fig=plt.figure(figsize=(3,2))
ax= fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes=plt.subplots(nrows=2,ncols=1,figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()#for not overlapping
```

```python
#to save figure in any format
fig.savefig('my_picture.jpg')
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,label='x')
ax.plot(y,x,label='y')
ax.set_title('Title')
ax.legend('loc=1')
```

#  plot Appearence

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='green',lw=5,alpha=0.4)#rgb hex code is also acceptable as color ,,, linewidth=thickness of the line,,,alpha=transparency of the line
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',lw=5,ls='steps')#ls=linestyle,lw=linewidth
#supported values  ls are '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted','steps'
```
