```python
import pandas as pd
from pandas_datareader import data,wb
import numpy as np
import datetime
%matplotlib inline
```

```python
start=datetime.datetime(2006,1,1)
end=datetime.datetime(2016,1,1)
```

```python
BAC=data.DataReader('BAC','yahoo',start,end)
```

```python
C=data.DataReader('C','yahoo',start,end)
```

```python
GS=data.DataReader('GS','yahoo',start,end)
```

```python
JPN=data.DataReader('JPN','yahoo',start,end)
```

```python
MS=data.DataReader('MS','yahoo',start,end)
```

```python
WFC=data.DataReader('WFC','yahoo',start,end)
```

```python
C.head()
```

```python
tickers=['BAC','C','GS','JPN','MS','WFC']
```

```python
bank_stocks=pd.concat([BAC,C,GS,JPN,MS,WFC],keys=tickers,axis=1)
```

```python
bank_stocks.head()
```

```python
bank_stocks.columns.names=['Bank Ticker','Stock Info']
```

```python
bank_stocks.head()
```

```python
bank_stocks['BAC']['Close'].max()
```

```python
for tick in tickers:
    print(tick,bank_stocks['BAC']['Close'].max())
```

```python
bank_stocks.xs(key='Close',axis=1,level='Stock Info').max()
```

```python
returns=pd.DataFrame()
```

```python
for tick in tickers:
    returns[tick+'Return']=bank_stocks[tick]['Close'].pct_changet_change()
```
